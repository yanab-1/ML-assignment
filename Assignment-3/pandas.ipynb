Task 1: Working with Series:-

1. Create a Pandas Series from the following list:
data = [25, 30, 35, 40, 45]
2. Write a Python program to:
-->Assign custom indices ['A', 'B', 'C', 'D', 'E'] to the Series.
-->Display the first three elements of the Series.
-->Calculate the mean, median, and standard deviation of the Series.

```python
import pandas as pd

# 1. Create a Pandas Series from the given list
data = [25, 30, 35, 40, 45]
custom_indices = ['A', 'B', 'C', 'D', 'E']

series = pd.Series(data, index=custom_indices)

# Display the Series
print("Pandas Series:")
print(series)
print()

# 2a. Display the first three elements
print("First three elements:")
print(series[:3])
print()

# 2b. Calculate mean, median, and standard deviation
print("Mean:", series.mean())
print("Median:", series.median())
print("Standard Deviation:", series.std())
```

Task 2: Creating and Inspecting DataFrames:-


1.  Create a Pandas DataFrame with the following data: 
Name Age Gender Marks 
Alice 20 Female  85 
Bob 22 Male 78 
Carol 19 Female  92 
David  21 Male 74 
Eve 20 Female  88 
 
2.  Write a Python program to: 
-->Display the first two rows of the DataFrame. 
-->Display the column names, data types, and summary statistics. 
-->Add a new column Passed that contains True if Marks >= 80 and False otherwise. 

```python
import pandas as pd

# Step 1: Create the DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'Age': [20, 22, 19, 21, 20],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female'],
    'Marks': [85, 78, 92, 74, 88]
}

df = pd.DataFrame(data)

# Step 2.1: Display the first two rows of the DataFrame
print("First two rows:")
print(df.head(2))

# Step 2.2: Display column names, data types, and summary statistics
print("\nColumn Names:")
print(df.columns.tolist())

print("\nData Types:")
print(df.dtypes)

print("\nSummary Statistics:")
print(df.describe(include='all'))

# Step 2.3: Add a new column 'Passed' where Marks >= 80
df['Passed'] = df['Marks'] >= 80
print("\nDataFrame with 'Passed' column:")
print(df)
```

Task 3: Data Selection and Filtering:-

1. Using the DataFrame from Task 2, write a Python program to:
-->Select and display the Name and Marks columns.
-->Filter and display records where Marks > 80.
-->Display the record of the student with the highest marks.


```python
import pandas as pd

# Recreate the DataFrame from Task 2
data = {
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'Age': [20, 22, 19, 21, 20],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female'],
    'Marks': [85, 78, 92, 74, 88]
}

df = pd.DataFrame(data)

# Add the 'Passed' column as in Task 2
df['Passed'] = df['Marks'] >= 80

# 1a. Select and display the Name and Marks columns
print("Name and Marks columns:")
print(df[['Name', 'Marks']])
print()

# 1b. Filter and display records where Marks > 80
print("Students with Marks > 80:")
print(df[df['Marks'] > 80])
print()

# 1c. Display the record of the student with the highest marks
max_marks = df['Marks'].max()
top_student = df[df['Marks'] == max_marks]
print("Student with the highest marks:")
print(top_student)
```

Task 4: Handling Missing Data
1. Modify the DataFrame from Task 2 by introducing missing values:
2. df.loc[1, 'Marks'] = None
3. df.loc[4, 'Age'] = None
4. Write a Python program to:
-->Identify missing values in the DataFrame.
-->Fill missing values in the Marks column with the column's mean.
-->Drop rows where the Age column has missing values.

```python
import pandas as pd

# Create the original DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'Age': [20, 22, 19, 21, 20],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female'],
    'Marks': [85, 78, 92, 74, 88]
}

df = pd.DataFrame(data)
print("Original DataFrame:-")
print(df)

# Introduce missing values
df.loc[1, 'Marks'] = None
df.loc[4, 'Age'] = None

# Identify missing values in the DataFrame
print("\nIdentify missing values in the DataFrame (True = missing):-")
print(df.isnull())

# Fill missing values in the Marks column with the column's mean
print("\nFilling missing values in the Marks column with the column's mean:-")
df['Marks'] = df['Marks'].fillna(df['Marks'].mean())
print(df)

# Drop rows where the Age column has missing values
print("\nDrop rows where the Age column has missing values:-")
df = df.dropna(subset=['Age'])
print(df)
```

Task 5: Grouping and Aggregation:-
1. Using the DataFrame from Task 2, write a Python program to:
-->Group the data by Gender and calculate the mean age and marks for each gender.
-->Count the number of students in each gender group.

```python
import pandas as pd

# Recreate the original DataFrame from Task 2
data = {
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'Age': [20, 22, 19, 21, 20],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female'],
    'Marks': [85, 78, 92, 74, 88]
}

df = pd.DataFrame(data)
print("Original DataFrame:-")
print(df)

# Group by Gender and calculate mean Age and Marks
print("\nMean Age and Marks by Gender:-")
grouped_mean = df.groupby('Gender')[['Age', 'Marks']].mean()
print(grouped_mean)

# Count number of students in each gender group
print("\nNumber of students in each Gender group:-")
gender_counts = df['Gender'].value_counts()
print(gender_counts)
```

Task 6: Reading and Writing Data
1. Write a Python program to:
-->Save the modified DataFrame from Task 4 to a CSV file named students_data.csv.
-->Read the CSV file into a new DataFrame.
-->Display the first five rows of the newly loaded DataFrame.

```python
import pandas as pd

# Create and modify the DataFrame (from Task 4)
data = {
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'Age': [20, 22, 19, 21, 20],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female'],
    'Marks': [85, 78, 92, 74, 88]
}

df = pd.DataFrame(data)

# Introduce missing values
df.loc[1, 'Marks'] = None
df.loc[4, 'Age'] = None

# Fill missing values in the Marks column with the mean
df['Marks'] = df['Marks'].fillna(df['Marks'].mean())

# Drop rows where Age is missing
df = df.dropna(subset=['Age'])

# 👉 Save the modified DataFrame to a CSV file
df.to_csv('students_data.csv', index=False)
print("DataFrame saved to 'students_data.csv'.")

# 👉 Read the CSV file into a new DataFrame
df_new = pd.read_csv('students_data.csv')

# 👉 Display the first five rows of the new DataFrame
print("\nFirst five rows of the newly loaded DataFrame:")
print(df_new)
```

Task 7: General
1. Download a sample dataset from Kaggle or use any public dataset.
2. Write a Python program to:
-->Load the dataset using Pandas.
-->Perform exploratory data analysis (EDA) by summarizing key statistics, checking for missing values, and visualizing data trends using Matplotlib or Seaborn.
-->Document your findings in Markdown or comments.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset using Seaborn
df = sns.load_dataset('iris')

# ---- 📌 Exploratory Data Analysis (EDA) ----

# 1. Display the first few rows
print("First 5 rows of the dataset:")
print(df.head())

# 2. Summary statistics
print("\nSummary Statistics:")
print(df.describe())

# 3. Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

# 4. Count of each species
print("\nNumber of samples per species:")
print(df['species'].value_counts())

# 5. Visualizations

# a) Pairplot to see relationships between features
sns.pairplot(df, hue='species')
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()

# b) Boxplot for Sepal Length across species
plt.figure(figsize=(8, 5))
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title("Boxplot of Sepal Length by Species")
plt.show()

# c) Correlation heatmap (Fixed using numeric_only=True)
plt.figure(figsize=(6, 4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
